```python
# Importing packages
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from difflib import SequenceMatcher as sm
from sklearn.preprocessing import LabelEncoder, StandardScaler, OneHotEncoder
from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV, cross_validate, KFold
from sklearn.metrics import f1_score, accuracy_score, classification_report
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC, LinearSVC
from sklearn.utils import resample
from sklearn.feature_selection import RFE
from xgboost import XGBClassifier
from imblearn.over_sampling import SMOTE, ADASYN
import category_encoders as ce
import os, datetime
import pickle as pkl

%matplotlib inline
```

```python
def set_occupation_category(col_value):
    occupation_categories = ["Tech-support", "Craft-repair", "Other-service", "Sales", "Exec-managerial", 
                             "Prof-specialty", "Handlers-cleaners", "Machine-op-inspct", "Adm-clerical", 
                             "Farming-fishing", "Transport-moving", "Priv-house-serv", "Protective-serv", "Armed-Force"]
    max_score = 0
    for category in occupation_categories:
        score = sm(None, category, col_value).ratio()
        if score > max_score:
            max_score = score
            occupation_category = category
    return occupation_category
```

```python
def set_workclass_category(col_value):
    workclass_categories = ["Private", "Self-emp-not-inc", "Self-emp-inc", "Federal-gov", 
                            "Local-gov", "State-gov", "Without-pay", "Never-worked"]
    max_score = 0
    for category in workclass_categories:
        score = sm(None, category, col_value).ratio()
        if score > max_score:
            max_score = score
            workclass_category = category
    return workclass_category
```

```python
def preprocess_data(filename, train = True):
    # Reading the CSV onto a dataframe
    df = pd.read_csv(filename)
    # ### There are no duplicates in the data
    df.drop_duplicates()
    # ### Preprocessing: Age
    # 
    # #### Dropping rows with zeroes in the age column because subsituting it with the mean age of the dataset would create a bias towards the mean and might falsify data.
    # Converting negative values of age to absolute value
    df["age"] = df["age"].apply(abs)
    # Dropping rows with value of zero in the age column
    df = df[df["age"]!=0]
    # ### Preprocessing: workclass
    # 
    # - The external library called as difflib in Python measures the similarity between words. We have a list of correct workclass categories.
    # - Each value in the unclean workclass column will be compared against the correct set of categories and the most similar category from the list will be imputed against the dirty value.
    # - Use of regular expressions has its limitations and thus, this method has been chosen.
    # Dropping null rows
    df.dropna(subset=["workclass"], inplace=True)
    # Dropping rows with "?"
    df = df[df["workclass"]!="?"]
    # Set the correct category for dirty values using similarity matching
    df["workclass"] = df["workclass"].apply(set_workclass_category)
    # ### Column: fnlwgt
    #
    # ### Preprocessing isn't required for fnlwgt as the numerical values seem to be consistent.
    # ### Column: education
    #
    # ### The values seem to be okay and no cleaning is required in this case. Thus, a single histogram is presented.
    # ### Column: education-num
    # 
    # - The education-num column seems to be a direct mapped value of education column. But there are some negative values which need to be mapped correctly using the education column.
    temp_df = df[["education","education-num"]].copy()
    temp_df["education_count"] = temp_df.groupby(["education", "education-num"])["education"].transform("count")
    temp_df.drop_duplicates(inplace=True)
    indices_with_max = temp_df.groupby(["education"])["education_count"].transform(max) == temp_df["education_count"]
    temp_df = temp_df.loc[indices_with_max]
    education_mapping = dict(zip(temp_df["education"], temp_df["education-num"]))
    # Mapping is done here
    df[["education", "education-num"]] = df[["education", "education-num"]].apply(lambda x: pd.Series((x["education"], education_mapping[x["education"]])), axis=1)
    # ### The histogram of education_num is very similar to that of education column because they are both the same
    # #### Thus, one of the columns can be dropped.
    df = df.loc[:, df.columns!="education"]
    # ### Column: marital-status
    # 
    # #### This column seems to be perfect and does not need any cleaning.
    # ### Preprocessing: occupation
    # 
    # - The external library called as difflib in Python measures the similarity between words. We have a list of correct workclass categories.
    # - Each value in the unclean occupation column will be compared against the correct set of categories and the most similar category from the list will be imputed against the dirty value.
    # - Use of regular expressions has its limitations and thus, this method has been chosen.
    # Dropping rows that have "?" in occupation column
    df = df[df["occupation"]!="?"]
    # Set the correct category for dirty values using similarity matching
    df["occupation"] = df["occupation"].apply(set_occupation_category)
    # ### Column: relationship
    # 
    # #### This column seems to be perfect and does not need any cleaning.
    # ### Column: race
    # 
    # #### This column seems to be perfect and does not need any cleaning.
    # ### Column: sex
    # 
    # #### This column seems to be perfect and does not need any cleaning.
    # ### Columns: capital-gain and capital-loss
    # 
    # #### Since most of the rows have a value of 0 in both these columns, the columns are going to be dropped.
    df = df.loc[:, df.columns!="capital-gain"]
    df = df.loc[:, df.columns!="capital-loss"]
    # ### Column: native-country
    # 
    # #### This column seems to be perfect and does not need any cleaning.
    if train:
        # ### Preprocessing: salary
        # 
        # - Rows with "?" in the salary column are dropped as the training/testing data cannot have null/erroneous value in the class variable.
        df = df[df["salary"]!="?"]
    # ### Column: hours-per-week
    # 
    # #### This a numerical column which does not need any preprocessing because the values seem okay and can be normalized.
    # Converting to CSV
    df.to_csv("{}__processed.csv".format(filename.split(".")[0]), index=False)
    return df
```

```python
def normalize(df, test_df, columns=[], custom=False):
    scaler = StandardScaler()
    for column in columns:
      if custom:
        max_column = max(df[column])
        df[column] = df[column]/max_column
        test_df[column] = test_df[column]/max_column
      else:
        df[column] = scaler.fit_transform(np.array(df[column]).reshape(-1, 1))
        test_df[column] = scaler.fit_transform(np.array(test_df[column]).reshape(-1, 1))
    return df, test_df
```

## Q1: Custom class to encode the categorical columns

```python
class CustomColumnEncoder():

  def __init__(self, df, test_df, le_columns=None, ohe_columns=None):
    self.df = df
    self.test_df = test_df
    self.ohe_columns = ohe_columns
    if ohe_columns:
      le_columns.extend(self.ohe_columns)
    self.le_columns = le_columns
    if le_columns:
      self.le()
    if ohe_columns:
      self.ohe()
    
  # One-hot encoding columns
  def ohe(self):
    df_columns = list(df.columns)
    self.ohe_indices = [df_columns.index(column) for column in self.ohe_columns]
    one_hot_encode = OneHotEncoder(sparse=False, categorical_features=self.ohe_indices)
    one_hot_encode.fit(self.df)
    self.df = one_hot_encode.transform(self.df)
    self.test_df = one_hot_encode.transform(self.test_df)
  
  # Label encoding columns
  def le(self):
    for column in self.le_columns:
      label_encode = LabelEncoder()
      label_encode.fit(self.df[column])
      self.df[column] = label_encode.transform(self.df[column])
      if column in list(self.test_df.columns):
        self.test_df[column] = label_encode.transform(self.test_df[column])
    
  # Binary encoding columns
  def be(self, columns=[]):
    self.be = ce.BinaryEncoder(cols=columns)
    self.df = self.be.fit_transform(self.df)
    return self.df

  # Get the Dataframe
  def get_df(self):
    return self.df

  def get_test_df(self):
    return self.test_df
```

### Encoding the columns for a linear model classifier

```python
cce = CustomColumnEncoder(df.copy(), test_df.copy(),
                          le_columns= ["salary"],
                          ohe_columns= ["workclass", "marital-status", "occupation", "relationship", "race", "sex", "native-country"])
linear_models_df, linear_models_test_df = cce.get_df(), cce.get_test_df()
```

### A linear model has certain weights associated with every feature. Label encoding values can induce bias towards certain attributes of a feature by assigning a higher value. Thus, to give equal importance to every attribute of a feature, the columns are one hot encoded. However, the education-num column is label encoded by default because it makes sense that certain people with a higher education with have a higher salary. Also, the order in which they are encoded already in "education-num" seems okay.

## Q5: Reasons for selecting the XGBoost classifier

- Firstly, the dataset is imbalanced and tree models generally tend to perform better in such cases.
- Secondly, XGBoost classifier is an ensemble that creates a strong classifier out of multiple weak learners sequentially. This is specifically helpful for this problem as the wrong classifications with every weak learner is rectified in the next sequential model built from it.
- Thirdly, on trying out various classification algorithms based on the evaluation metric of f1_macro, XGBoost performed better.

## Q5: Steps taken to fine tune the XGBoost classifier

- Firstly, default parameters of XGBoost were tried.
- Then, the hyperparameters of the XGBoost classifier was fine tuned by a GridSearchCV model which takes all possible combinations of the given values and runs a cross validation on it. The end result is a classifier with the best hyperparameters.
- Finally, a feature selection method such as relief feature selection was used in order to maximise the f1_macro. This method takes into account only the top n features that are very important for classification. The n was decided by brute force method and n=8 fetched the best results on the validation set.
